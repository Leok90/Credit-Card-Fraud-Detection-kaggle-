### 목적
- 기회탐지 (Chance discovery, Positive Anomaly) 새로움 탐지 (Novelty Detection)

- 오류탐지 (Fault Discovery, Negative Anomaly) 노이즈 제거 (Noise Removal)

### 입력 데이터의 특성
1. Time-Series(sequential) vs Static
2. Univeriate vs Multivariate
3. Data Type(Binary / Categorical / Continuous / Hybrid)
4. Relational vs Independent
5. Well-known or not (rule existing or not)
(기존 룰의 적용이 가능할 만큼 잘 알려진/알려져 있지 않은)

### Point anomaly
축적된 시간동안 정적인 점분포에 초점</br>
ex) Gaussian distribution, Clustering, Classification

### Contextual anomaly
시계열과 같은 동적인 특성에 초점</br>
ex) Gaussian Process Regression, S-H-ESD ...)

### Anomaly 종류
#### Global Outlier : 직관적으로 떠오르는 Normal 상태와 구분되는 특이값
- 일반적인 관측치들과 상당히 동떨어진 관측치들을 의미
- 얼마나 떨어져 있는가를 어떻게 측정할 것인가?

#### Contextual Outlier : 특정 부분에서 다른 부분과는 다른 양상을 띄는 이상치 관측된 값들의 context를 어떻게 설정하는지가 중요



### 머신러닝 지도학습(Supervised Learning)을 이용한 이상치 탐지 접근 방식
이상치가 있는 데이터를 학습</br>
회귀/SVM/Decision Tree/딥러닝 등의 알고리즘을 사용하여 예측(Prediction)
- 이상치는 극소수에 불과하여 실제 트레이닝 시킬 데이터가 충분치 아니함
- 종종 '이상치'로 규명한 것이 매우 애매하게 다른 '정상치'와 구분이 어려울 수 있음

### 머신러닝 비지도학습이 이상탐지에 적용이 어려운 이유
1. 이상탐지는 판별의 문제이기 때문에 어떻게 이상한 지 아닌지를 판별해야 함
2. 비지도학습은 정답이 주어지고 유사한 것을 찾는 문제를 해결하는 것은 일반적으로 부적합함
3. 분리된 군집들 중에 이상치만 모여있는 모델을 만드는 것은 매우 어려움

### 이상탐지 ML 알고리즘
1. 규칙유도
- 데이터의 패턴을 찾아 IF-THEN 형태의 규칙으로 정의하는 방법
- 데이터의 일부 또는 전체를 표현하는 규칙을 반복적으로 찾아 나가면서 규칙 집합을 생헝하고, 과적합을 완화시키기 위해 프루닝(가지치기) 과정을 거침

2. Random Forest(ML)
- 단일 학습 알고리즘의 성능보다 더 높은 성능을 얻기 위해 다중 학습 알고리즘을 사용하는 앙상블 학습방법의 일종
- 다수의 의사결정나무 알고리즘으로부터의 평균치를 이용하여 동작

3. 서포트벡터머신(SVM, ML)
- 딥러닝 활성화 이전시대에 가장 많이 쓰던 분류 알고리즘으로 기본적으로는 데이터를 두 개의 그룹으로 분류
- 이 때 두개의 그룹으로 데이터를 분류하는 최적의 기준선을 찾는 것이 알고리즘의 목적

4. 자기조직화맵(SOM, ML)
- 비지도 학습 신경망 모델의 유형 중 하나로 정보 집합을 그래프로 표현하여 분석
- 유사한 패턴을 가진 정보들을 클러스터링

5. 히든 마르코프 모델(HMM, ML)
- 음성, 필기체 인식 등과 같이 시간에 따라 변화하는 것에서 패턴을 인식하는데 유용하게 사용
- 확률이론을 기반으로 하며 이전 상태와 현재 상태의 값을 이용해 미래의 값을 예측하는 방법
- 고려하는 상태의 수에 따라 알고리즘 복잡도가 증가

6. 유전 알고리즘
- 유전학에서 다윈의 진화론 개념을 이용한 것으로 유전자가 세대를 반복하면서 자연환경에 적응할 수 있는 적합한 형태로 변이
- 살아남는 것처럼 해결하고자 하는 문제의 가능한 해들을 대상으로 선택, 교차, 변이 등을 수행함으로써 가능한 최적의 해를 찾아내는 방법

7. 딥러닝
- 인간 신경망 구조에서 착안한 알고리즘으로 입력층, 출력층, 다수의 은닉층을 구성하여 학습하는 방식
- CNN, RNN 등이 존재

### Anomaly Detection Use Case
#### 비즈니스 분석</br>
매출이 이전에 비해 떨어진 판매채널 또는 주물량이 이전에 비해 증가한 거래처
- 다른 경쟁사로 이탈할 징후가 보이는 쇼핑몰 또는 서비스의 고객
- 평소와 다른 공간 이동 패턴을 보이는 차량 또는 모바일 디바이스 이용자
</br>

#### 컴퓨터 시스템 관리
이전에 없던 접속 IP 주소로부터 시스템 접근량의 증가(DDOS 징후)
- 서버, 스토리지, 네트워크 장비 등의 이상징후 탐지
- 리소스 사용량이 이전에 비해 급증한 시스템 탐지
</br>

#### 온라인 서비스 관리
다른 사용자와 다른 이동패턴이나 접근 패턴을 보이는 사용자
- 이전에 비해 증가한 로그인 실패 기록
- 이전에 발생하지 않은 에러코드의 기록 또는 더 빈번해진 기록

</br>
#### 보안
사람의 접근이 거의 없는 지역에 이전보다 다소 증가한 통행량
- 사내 기밀문서 유출 횟수가 상대적으로 증가한 직원
- 임계치를 넘지는 않는 범위내에서 이전보다 증가한 보안 접근실패


### FDS의 구조 (Fraud Detection System)
1. 룰 기반 FDS
- 정탐 : 기존 생성한 룰 조합 내 패턴과 일치 -> 이상거래의 확률
- 오탐 : 기존 생성한 룰 조합 내 패턴과 일치 -> 정상거래의 확률
- 미탐 : 기존 생성한 룰 조합 외 신규 패턴 -> 탐지 불가
</br>
정탐/오탐 -> 사전 학습패턴으로 대응 가능성 있음
</br> 미탐 -> 신규 사기패턴에 대해서 대응 불가
</br> 수집 인자가 많을 수록 대응 불가
